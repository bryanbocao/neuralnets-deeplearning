```python
import numpy as np
import random
import sys

data = np.loadtxt('assign1_data.txt')
#print data

x1 = data[:, 0]
print "x1:", x1

x2 = data[:, 1]
print "x2:", x2

y = data[:, 2]
#print "y:", y

z = data[:, 3]
#print "z:", z
```

```python
# initialize weights
w1 = random.random() * 3
w2 = random.random() * 3
b = random.random() * 3

print "w1:\t", w1
print "w2:\t", w2
print "b:\t", b
```

```python
# online
lr = 0.0001 # learning rate
epoch = 0
# training set size : validation set size = 3 : 1
training_set_index = 0
size = len(x1)
validation_set_index = 0
validation_set_index_step = size / 4
pre_validation_error = sys.maxint
curr_validation_error = 0
while pre_validation_error >= curr_validation_error: # early stopping when the validation error is greater than its previous one
#while True:
    
    validation_set_index += validation_set_index_step
    validation_set_index %= size
    
    x1_validation = x1[validation_set_index : validation_set_index + validation_set_index_step]
    x2_validation = x2[validation_set_index : validation_set_index + validation_set_index_step]
    y_validation = y[validation_set_index : validation_set_index + validation_set_index_step]
    
    '''
    if validation_set_index == 0:
        training_set_index = validation_set_index_step
        training_set_index_end = size
        x1_training = x1[training_set_index : training_set_index_end]
        x2_training = x2[training_set_index : training_set_index_end]
        y_training = y[training_set_index : training_set_index_end]
    elif validation_set_index + validation_set_index_step == size:
        training_set_index = 0
        training_set_index_end = validation_set_index
        x1_training = x1[training_set_index : training_set_index_end]
        x2_training = x2[training_set_index : training_set_index_end]
        y_training = y[training_set_index : training_set_index_end]
    else:
        training_set_index0 = 0
        training_set_index_end0 = validation_set_index
        training_set_index1 = validation_set_index + validation_set_index_step
        x1_training = x1[training_set_index0 : training_set_index_end0]
        print "x1_training:\t", x1_training
        x1_training.extend(x1[training_set_index1 :])
        x2_training = x2[training_set_index0 : training_set_index_end0]
        x2_training.extend(x2[training_set_index1 :])
        y_training = y[training_set_index0 : training_set_index_end0]
        y_training.extend(y[training_set_index1 :])
    
   
    # training
    for x1i, x2i, yi in zip(x1_training, x2_training, y_training):
        w1 = w1 + lr * (yi - (x1i * w1 + x2i * w2 + b)) * x1i
        w2 = w2 + lr * (yi - (x1i * w1 + x2i * w2 + b)) * x2i
        b = b + lr * (yi - (x1i * w1 + x2i * w2 + b))
        
    pre_validation_error = curr_validation_error
    # validation
    for x1i, x2i, yi in zip(x1_validation, x2_validation, y_validation):
        curr_validation_error += abs(yi - (x1i * w1 + x2i * w2 + b))
        
    epoch += 1
    # print "w1:\t", w1, "\tw2:\t", w2, "\tb:\t", b, "\tepoch:\t", epoch
    '''
   
    # training
    for x1i, x2i, yi in zip(x1, x2, y):
        if x1i not in zip(x1_validation) and x2i not in zip(x2_validation) and yi not in zip(y_validation):
            w1 = w1 + lr * (yi - (x1i * w1 + x2i * w2 + b)) * x1i
            w2 = w2 + lr * (yi - (x1i * w1 + x2i * w2 + b)) * x2i
            b = b + lr * (yi - (x1i * w1 + x2i * w2 + b))

    if epoch >= 4 and epoch % 4 == 0:
        pre_validation_error = curr_validation_error
        print "epoch:\t", epoch, "\t curr_validation_error:\t", curr_validation_error
        curr_validation_error = 0
        
    # validation
    for x1i, x2i, yi in zip(x1_validation, x2_validation, y_validation):
        curr_validation_error += abs(yi - (x1i * w1 + x2i * w2 + b))
                
    epoch += 1

print "w1:\t", w1, "\tw2:\t", w2, "\tb:\t", b
    
```
